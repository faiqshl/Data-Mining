```python
#library untuk manipulasi data
import pandas as pd
#library untuk naive bayes
from sklearn.naive_bayes import GaussianNB
#library untuk mengubah data kategorikal menjadi numeric
from sklearn.preprocessing import LabelEncoder
```

```python
#import dataset
df = pd.read_csv('diagnosis peradangan akut.csv')
```

```python
df
```

```python
#rename kolom
df.columns = ['suhu_tubuh','mual','nyeri_pinggang','BAK_menerus','MP','infeksi_uretra','radang_kandungkemih','nefritis']
```

```python
df
```

# 10 data

```python
#mengubah , menjadi . pada kolom suhu_tubuh
df['suhu_tubuh'] = df['suhu_tubuh'].apply(str)
df['suhu_tubuh'] = df['suhu_tubuh'].str.replace(',','.').apply(float)
df
```

```python
#menggunakan data baris 0-9
data_10 = df[0:10]
data_10
```

```python
#cleaning data
data_10.isna().sum()
```

```python
#seleksi data
data_10 = data_10.loc[:, ['suhu_tubuh','BAK_menerus','MP','infeksi_uretra','radang_kandungkemih']]
data_10
```

```python
#transformasi data
for index, row in data_10.iterrows():
    if row['suhu_tubuh'] > 37.5:
        data_10.loc[index, 'suhu_tubuh'] = 'tinggi'
    elif row['suhu_tubuh'] >= 36.5:
        data_10.loc[index, 'suhu_tubuh'] = 'normal'
    else:
        data_10.loc[index, 'suhu_tubuh'] = 'rendah'
        
data_10
```

```python
#klasifikasi (naive bayes)
#mengubah kategorikal menjadi numerik (0,1,2,...)
number = LabelEncoder()
data_10['suhu_tubuh'] = number.fit_transform(data_10['suhu_tubuh'])
data_10['BAK_menerus'] = number.fit_transform(data_10['BAK_menerus'])
data_10['MP'] = number.fit_transform(data_10['MP'])
data_10['infeksi_uretra'] = number.fit_transform(data_10['infeksi_uretra'])
data_10['radang_kandungkemih'] = number.fit_transform(data_10['radang_kandungkemih'])
```

```python
#menentukan atribute dan target 
attributes = ['suhu_tubuh','BAK_menerus','MP','infeksi_uretra']
target = ['radang_kandungkemih']
```

```python
#menggunakan naive bayes
model = GaussianNB()
```

```python
atr = data_10[attributes]
target = data_10[target]
```

```python
model.fit(atr, target)
```

```python
#skor akurasi dari prediksi pada data testing
model.score(atr, target)
```

```python
data_10
```

keterangan :

suhu_tubuh 0 = normal, 1 = rendah

BAK_menerus 0 = no, 1 = yes

MP 0 = no, 1 = yes

infeksi_uretra 0 = no, 1 = yes

radang_kandungkemih 0 = no, 1 = yes

```python
#prediksi data
#rendah, yes, no, no
print (model.predict([[1,1,0,0]]))
```

```python
#prediksi menggunakan atribut yang ada di data asli /data awal
print (model.predict([[1,0,0,0]]))
print (model.predict([[1,1,1,1]]))
print (model.predict([[1,0,0,0]]))
print (model.predict([[1,1,1,1]]))
print (model.predict([[1,0,0,0]]))
print (model.predict([[1,0,0,0]]))
print (model.predict([[1,1,1,1]]))
print (model.predict([[1,0,0,0]]))
print (model.predict([[1,1,1,1]]))
print (model.predict([[0,1,1,1]]))
```

```python
#pengujian dengan confusion matrix
from sklearn.metrics import confusion_matrix, classification_report 
```

```python
prediksi = model.predict(atr)
confusion_matrix(prediksi, target)
```

```python
print(classification_report(prediksi, target))
```

```python
std_suhu_tubuh = (statistics.stdev(df.suhu_tubuh))
```

```python
T_suhuTubuh_max = mean_suhu_tubuh + (2 * std_suhu_tubuh)
T_suhuTubuh_min = mean_suhu_tubuh - (2 * std_suhu_tubuh)
```

```python
df[(df.suhu_tubuh > T_suhuTubuh_max) | (df.suhu_tubuh < T_suhuTubuh_min)].index
```

```python
#seleksi data
df = df.loc[:, ['suhu_tubuh','BAK_menerus','MP','infeksi_uretra','radang_kandungkemih']]
df
```

```python
#transformasi data
for index, row in df.iterrows():
    if row['suhu_tubuh'] > 37.5:
        df.loc[index, 'suhu_tubuh'] = 'tinggi'
    elif row['suhu_tubuh'] >= 36.5:
        df.loc[index, 'suhu_tubuh'] = 'normal'
    else:
        df.loc[index, 'suhu_tubuh'] = 'rendah'
        
df
```

```python
#klasifikasi (naive bayes)
#mengubah kategorikal menjadi numerik (0,1,2,...)
number = LabelEncoder()
df['suhu_tubuh'] = number.fit_transform(df['suhu_tubuh'])
df['BAK_menerus'] = number.fit_transform(df['BAK_menerus'])
df['MP'] = number.fit_transform(df['MP'])
df['infeksi_uretra'] = number.fit_transform(df['infeksi_uretra'])
df['radang_kandungkemih'] = number.fit_transform(df['radang_kandungkemih'])
```

```python
#menentukan atribute dan target 
attributes1 = ['suhu_tubuh','BAK_menerus','MP','infeksi_uretra']
target1 = ['radang_kandungkemih']
```

```python
x = df[attributes1]
y = df[target1]
```

```python
#membagi data training dan test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
print("ukuran data train, dan kelas train : ", x_train.shape, y_train.shape)
print("ukuran data test, dan kelas test : ", x_test.shape, y_test.shape)
```

```python
#menggunakan naive bayes
model1 = GaussianNB()
```

```python
model1.fit(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
#uji dengan confusion matriks
prediksi1 = model1.predict(x_test)
confusion_matrix(prediksi1, y_test)
```

```python
#menampilkan hasil precision, reacl, f-1score, support, beserta akurasi score nya
print(classification_report(prediksi1, y_test))
```
